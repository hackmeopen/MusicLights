```python
channel = 0

# Function to send all MIDI messages to FluidSynth
def send_midi_to_fluidsynth(message):
    
    if message.type == 'note_on':
        synth.noteon(message.channel, message.note, message.velocity)
        if message.channel == channel:
            control_lights(message.note, 1)
    elif message.type == 'note_off':
        synth.noteoff(message.channel, message.note)
        if message.channel == channel:
            control_lights(message.note, 0)
    elif message.type == 'control_change':
        synth.cc(message.channel, message.control, message.value)
    elif message.type == 'program_change':
        synth.program_change(message.channel, message.program)
    # You can add other MIDI message types as needed

# Function to play MIDI file events
def play_midi_file(midi_file_path):
    # Open the MIDI file with Mido
    mid = mido.MidiFile(midi_file_path)

    # Create a MIDI input for receiving and processing events
    for message in mid.play():
        send_midi_to_fluidsynth(message)
        
        # # Sleep to match the tempo of the MIDI file
        # time.sleep(message.time)
```

```python
import pcf8574_io

#start the i2C with the address
p1 = pcf8574_io.PCF(0x27)
p1.pin_mode("p0", "OUTPUT")
p1.pin_mode("p1", "OUTPUT")
p1.pin_mode("p2", "OUTPUT")
p1.pin_mode("p3", "OUTPUT")
p1.pin_mode("p4", "OUTPUT")
p1.pin_mode("p5", "OUTPUT")
p1.pin_mode("p6", "OUTPUT")
p1.pin_mode("p7", "OUTPUT")

note_light_dict = {}
next_light_index = 0
num_lights = 5

def control_lights(note, on_off):

    global next_light_index
    if on_off:
        command = "LOW"
    else:
        command = "HIGH"

    #check to see if the note is already in the dictionary
    if note in note_light_dict:
        light = note_light_dict[note]
        p1.write(light, command)
    else:
        light = 'p' + str(next_light_index)
        next_light_index += 1
        if next_light_index == num_lights:
            next_light_index = 0
        note_light_dict[note] = light
```

```python
import mido
import time
import fluidsynth

# Initialize FluidSynth with a SoundFont
soundfont_path = 'Nintendo_64_ver_2.0.sf2'
synth = fluidsynth.Synth()
synth.start()
sfid = synth.sfload(soundfont_path)  # Load the SoundFont
synth.program_select(0,sfid, 0, 0)  # Select bank, soundfont, preset, and channel
synth.maxpolyphony = 128

midi_file_path = 'jingle_bells_bb.mid'
play_midi_file(midi_file_path)
```
